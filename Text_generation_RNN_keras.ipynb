```python
def read_file(filepath):
    with open(filepath) as f:
        str_text = f.read()
    return str_text
```

```python
import spacy
```

```python
import en_core_web_sm
nlp = en_core_web_sm.load(disable=['parser','tagger','ner'])
```

```python
nlp.max_length= 1198623
```

```python
def separate_punct(doc_text):
    return[token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()']
```

```python
d = read_file('moby_dick_four_chapters.txt')
```

```python
tokens = separate_punct(d)
```

```python
len(tokens)
```

```python
#pass 25 words--> predict 26
```

```python
train_len =25 +1
text_seq = []
for i in range(train_len,len(tokens)):
    seq = tokens[i-train_len:1]
    text_seq.append(seq)
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(text_seq)
```
